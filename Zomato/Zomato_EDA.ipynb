```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
new_df = pd.merge(df,df_country, on='Country Code', how='left')
```

```python
new_df.head()
```

```python
new_df.Country.value_counts()
```

```python
country_name = new_df.Country.value_counts().index
```

```python
country_name
```

```python
country_counts = new_df.Country.value_counts().values
```

```python
plt.pie(x = country_counts[:3], labels = country_name[:3], autopct = '%1.2f%%')
plt.show()
```

# insights: 
- zomatos max records are from India then USA followed by United Kingdom

```python
ratings = new_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating counts'})
ratings
```

# insights 
- there are 6 types of ratings given when the rating is:
1. From 4.5 to 4.9 is Excellent
2. From 4.0 to 4.4 is Very Good 
3. From 3.5 to 3.9 is Good
4. From 2.5 to 3.4 is  average
5. From 1.8 to 2.4 is poor

```python
ratings.head()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='Aggregate rating', y ='Rating counts', hue = 'Rating color',data = ratings)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='Aggregate rating', y ='Rating counts', hue = 'Rating color',data = ratings, palette = ['b','red','orange','yellow','green','green'])
plt.show()
```

# insights
- Count of no ratings very high
- max no of people gave between 2.5 to 3.4

```python
sns.countplot(x='Rating color', data=ratings, palette=['b','red','orange','yellow','green','green'])
plt.show()
```
