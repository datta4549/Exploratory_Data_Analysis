```python
df['publishedAt'] = pd.to_datetime(df['publishedAt'], format = '%Y-%m-%dT%H:%M:%SZ', errors='coerce')
```

```python
df['trending_date'] = pd.to_datetime(df['trending_date'], format = '%Y-%m-%dT%H:%M:%SZ', errors='coerce')
```

```python
week_ago_condition=(df['trending_date'] >= week_ago) & (df['trending_date'] <= today)
IND_trending_videos_week_ago = df[week_ago_condition]
```

##Exploratory Analysis and Visualization

Likes vs views relations

```python
plt.figure(figsize=(15,7))

sns.lmplot(x='view_count', y='likes', data=df)
plt.show()
```

#top 10 most viewed videos this week

```python
top_10_viewed_vid = IND_trending_videos_week_ago.groupby(['title']).max().sort_values('view_count',ascending=False).loc[:,'view_count'][:10]
```

```python
top_10_vid = top_10_viewed_vid.reset_index()
```

```python
top_10_vid
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='view_count', y='title', data=top_10_vid)
plt.title('Top 10 Most Viewed Videos Last Week')
plt.show()
```

Top 10 most liked video of this week

```python
top_10_most_liked = IND_trending_videos_week_ago.groupby(['title']).max().sort_values('likes',ascending=False).loc[:,'likes'][:10].reset_index()
top_10_most_liked
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='likes', y='title', data=top_10_most_liked)
plt.title('Top 10 Most Liked Videos Last Week')
plt.show()
```

Top 10 Most disliked video

```python
top_10_most_disliked = IND_trending_videos_week_ago.groupby(['title']).max().sort_values('dislikes',ascending=False).loc[:,'dislikes'][:10].reset_index()
top_10_most_disliked
```

```python
plt.figure(figsize=(10,8))

sns.barplot(x='dislikes', y = 'title', data=top_10_most_disliked)
plt.title("Top 10 Most Disliked Video")
plt.show()
```

Top 10 Trending Channel This Week

```python
top_10_trending_chnl = IND_trending_videos_week_ago.groupby(['channelTitle']).size().sort_values(ascending=False).head(10).reset_index()
top_10_trending_chnl.rename(columns={0:'counts'}, inplace = True)
```

```python
plt.figure(figsize=(10,8))

sns.barplot(x='counts', y = 'channelTitle', data=top_10_trending_chnl)
plt.title("Top 10 Trending Channels This week")
plt.show()
```
